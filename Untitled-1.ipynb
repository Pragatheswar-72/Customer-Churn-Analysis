```python
# ======================================================
# CUSTOMER CHURN ANALYSIS PROJECT
# Tools: Python, Pandas, NumPy, Matplotlib, Seaborn, Statistics
# ======================================================

# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from scipy.stats import ttest_ind

# ======================================================
# LOAD DATASET
# ======================================================

df = pd.read_csv("Telco_Customer_Churn.csv")

print("=" * 50)
print("DATASET SHAPE")
print("=" * 50)

print(df.shape)

print("\n")

print("=" * 50)
print("FIRST 5 ROWS")
print("=" * 50)

print(df.head())

# ======================================================
# DATA INFORMATION
# ======================================================

print("\n")
print("=" * 50)
print("DATASET INFO")
print("=" * 50)

print(df.info())

print("\n")

print("=" * 50)
print("MISSING VALUES")
print("=" * 50)

print(df.isnull().sum())

# ======================================================
# DATA CLEANING
# ======================================================

print("\nCleaning Data...")

# Remove Duplicates
df.drop_duplicates(inplace=True)

# Convert TotalCharges

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

# Fill Missing Values

df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)

print("\nMissing Values After Cleaning")

print(df.isnull().sum())

# ======================================================
# DESCRIPTIVE STATISTICS
# ======================================================

print("\n")
print("=" * 50)
print("DESCRIPTIVE STATISTICS")
print("=" * 50)

print("Mean Monthly Charges:")
print(df["MonthlyCharges"].mean())

print("\nMedian Monthly Charges:")
print(df["MonthlyCharges"].median())

print("\nMode Monthly Charges:")
print(df["MonthlyCharges"].mode()[0])

print("\nStandard Deviation:")
print(df["MonthlyCharges"].std())

print("\nMinimum:")
print(df["MonthlyCharges"].min())

print("\nMaximum:")
print(df["MonthlyCharges"].max())

# ======================================================
# CHURN RATE
# ======================================================

print("\n")
print("=" * 50)
print("CUSTOMER CHURN RATE")
print("=" * 50)

churn_rate = (
    df["Churn"]
    .value_counts(normalize=True)
    * 100
)

print(churn_rate)

# ======================================================
# CHURN DISTRIBUTION
# ======================================================

plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="Churn"
)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")

plt.show()

# ======================================================
# CONTRACT TYPE VS CHURN
# ======================================================

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)

plt.title("Contract Type vs Churn")

plt.xticks(rotation=20)

plt.show()

# ======================================================
# GENDER VS CHURN
# ======================================================

plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="gender",
    hue="Churn"
)

plt.title("Gender vs Churn")

plt.show()

# ======================================================
# MONTHLY CHARGES DISTRIBUTION
# ======================================================

plt.figure(figsize=(8,5))

sns.histplot(
    df["MonthlyCharges"],
    kde=True
)

plt.title("Monthly Charges Distribution")

plt.show()

# ======================================================
# TENURE DISTRIBUTION
# ======================================================

plt.figure(figsize=(8,5))

sns.histplot(
    df["tenure"],
    kde=True
)

plt.title("Customer Tenure Distribution")

plt.show()

# ======================================================
# BOXPLOT FOR OUTLIERS
# ======================================================

plt.figure(figsize=(8,4))

sns.boxplot(
    x=df["MonthlyCharges"]
)

plt.title("Outlier Detection - Monthly Charges")

plt.show()

# ======================================================
# CHURN FLAG CREATION
# ======================================================

df["Churn_Flag"] = df["Churn"].map({
    "Yes": 1,
    "No": 0
})

# ======================================================
# CORRELATION ANALYSIS
# ======================================================

print("\n")
print("=" * 50)
print("CORRELATION ANALYSIS")
print("=" * 50)

corr_columns = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges",
    "Churn_Flag"
]

correlation_matrix = df[corr_columns].corr()

print(correlation_matrix)

# ======================================================
# CORRELATION HEATMAP
# ======================================================

plt.figure(figsize=(8,6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()

# ======================================================
# HYPOTHESIS TESTING
# ======================================================

print("\n")
print("=" * 50)
print("HYPOTHESIS TESTING")
print("=" * 50)

# Null Hypothesis:
# Monthly charges do not differ
# between churned and retained customers

churned = df[
    df["Churn"] == "Yes"
]["MonthlyCharges"]

retained = df[
    df["Churn"] == "No"
]["MonthlyCharges"]

t_statistic, p_value = ttest_ind(
    churned,
    retained
)

print("T Statistic:", t_statistic)
print("P Value:", p_value)

if p_value < 0.05:
    print("\nReject Null Hypothesis")
    print("Monthly charges significantly affect churn.")
else:
    print("\nFail to Reject Null Hypothesis")
    print("No significant difference found.")

# ======================================================
# CUSTOMER SEGMENTATION
# ======================================================

print("\n")
print("=" * 50)
print("HIGH VALUE CUSTOMERS")
print("=" * 50)

high_value_customers = df[
    df["MonthlyCharges"] > 80
]

print("Total High Value Customers:")
print(high_value_customers.shape[0])

# ======================================================
# TOP CONTRACT TYPES
# ======================================================

print("\n")
print("=" * 50)
print("CONTRACT ANALYSIS")
print("=" * 50)

contract_analysis = pd.crosstab(
    df["Contract"],
    df["Churn"]
)

print(contract_analysis)

# ======================================================
# INTERNET SERVICE ANALYSIS
# ======================================================

plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="InternetService",
    hue="Churn"
)

plt.title("Internet Service vs Churn")

plt.xticks(rotation=15)

plt.show()

# ======================================================
# PAYMENT METHOD ANALYSIS
# ======================================================

plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="PaymentMethod",
    hue="Churn"
)

plt.title("Payment Method vs Churn")

plt.xticks(rotation=45)

plt.show()

# ======================================================
# EXPORT CLEANED DATA
# ======================================================

df.to_csv(
    "cleaned_customer_churn.csv",
    index=False
)

print("\nCleaned dataset exported successfully.")

# ======================================================
# FINAL BUSINESS INSIGHTS
# ======================================================

print("\n")
print("=" * 50)
print("PROJECT SUMMARY")
print("=" * 50)

print("""
1. Calculated customer churn rate.

2. Performed descriptive statistics:
   - Mean
   - Median
   - Mode
   - Standard Deviation

3. Conducted exploratory data analysis.

4. Performed correlation analysis.

5. Conducted hypothesis testing using T-Test.

6. Identified high-value customers.

7. Exported cleaned dataset.

8. Generated business insights for retention strategies.
""")

```
