```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import ttest_ind

%matplotlib inline
```

```python
df = pd.read_csv("Telco_Customer_Churn.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)

df.isnull().sum()
```

```python
df.describe()
```

```python
df["Churn"].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Churn")
plt.title("Customer Churn Distribution")
plt.show()
```

```python
pd.crosstab(
    df["Contract"],
    df["Churn"]
)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)

plt.title("Contract Type vs Churn")
plt.show()
```

```python
df["Churn_Flag"] = df["Churn"].map({
    "Yes":1,
    "No":0
})

corr = df[
    ["tenure",
     "MonthlyCharges",
     "TotalCharges",
     "Churn_Flag"]
].corr()

corr
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```

```python
churned = df[df["Churn"]=="Yes"]["MonthlyCharges"]

retained = df[df["Churn"]=="No"]["MonthlyCharges"]

t_stat, p_value = ttest_ind(
    churned,
    retained
)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```

```python
print("1. Month-to-Month customers churn the most.")
print("2. Higher monthly charges increase churn risk.")
print("3. Long-tenure customers are more loyal.")
print("4. Contract type affects retention.")
```
